## Probability calculation for a continuous random variables

The length of a human pregnancy is normally distributed with mean of 272 days with a standard deviation of 9 days
1.State the random variable
2.Find the probability of a pregancy lasting less than 250 days.
3.Find the probability that a pregnancy lasts between 265 and 280 days.

Solutions
1.RV - length of Human pregnancy

```python
from scipy import stats
```

2. Find the probability of a pregnacy lasting more than 280 days.

```python
1 - stats.norm.cdf(x=280,loc=272,scale=9)
#x - value, loc - mean, scale - std
```

3.Find the probability that a pregnancy lasting less than 250 days.

```python
stats.norm.cdf(x=250,loc=272,scale=9)
```

3.Find the probability that a pregnancy lasts between 265 and 280 days.

```python
stats.norm.cdf(x=280,loc=272,scale=9) - stats.norm.cdf(x=265,loc=272,scale=9)
```

K-Means Clustering(Non Hierarchical clustering)

Steps involved in Kmeans clustering

1.Define Kvalue - no of clusters. And based on the k value, divide the entire dataset in k divisions
2.Compute the centroid for each group.
3.Findout the distance from each datapoint to the cluster centroids.
4.Move the datapoints to the nearest cluster centroid.
5.Due to this repostioning, we have to recompute the centroid.
6.Repeat the 3-5, until you don't need to move any datapoint to anyother clusters.

GOAL - "All the datapoints should have minimum distance from its cluster centroids."
Euclideon distance

##K means

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
Univ = pd.read_csv("Universities.csv")
Univ
```

```python
#Standardization function
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_Univ_df = scaler.fit_transform(Univ.iloc[:, 1:])
print(scaled_Univ_df)
```

```python
#How to find optimum number of cluster
#The Kmeans algorithm aims to choose centroids that minimize the inertia, or within cluster sum of squares criterian
```

```python
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=12)
    kmeans.fit(scaled_Univ_df)
    wcss.append(kmeans.inertia_)
```

wcss

```python
from sklearn.cluster import KMeans
clusters_new = KMeans(n_clusters=3, random_state=12)
clusters_new.fit(scaled_Univ_df)
```

```python
clusters_new.labels_
```

```python
#Assign clusters to the data set
Univ['clusterid_new']=clusters_new.labels_
```

```python
Univ
```

```python
clusters_new.cluster_centers_
```

```python
Univ.drop('Univ', axis=1).groupby('clusterid_new').agg(['mean'])
```

```python
Univ[Univ['clusterid_new']==0]
Univ
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('WCSS')
plt.show()
```
